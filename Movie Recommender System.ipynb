```python
import ast      # convert string to normal variable
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L     
```

```python
convert(movies.iloc[0].genres)
```

```python
movies['genres']
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# We have seperate convert because we want to only get 3 dictionary i.e only 3 casts
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else :
            break
    return L     
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def fetch_director_from_crew(obj):
    L = []
    
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L     
```

```python
movies['crew'] = movies['crew'].apply(fetch_director_from_crew)
```

```python
movies['overview'][0]  # this is overview of first movie which is a string
```

```python
#because it is a string so we convert it into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
movies['genres'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Stemming  
#which makes the word dancing, dance, danced, dancer as same word danc
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def my_stem(text):
    y = []
    for i in text.split():
         y.append(ps.stem(i))
            
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(my_stem)
```

```python
new_df['tags']
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english') # take 5000 most repeting words and compare it with all our 5000 movies but don't use stop words
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
list(enumerate(similarity[0]))       # we need the index of the similarity of 1st movie with 5000 movies so use enumerate function
```

```python
sorted(list(enumerate(similarity[0])),reverse=True)    # but this sorts according to the index number not similarity 
```

```python
#so we need to sort in descending order according to cosine index, so we use key = x[1]
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
# we need top values from and since we dont take the 1st value because it is the cosine similarity with itself 
# top 5 movies with their index
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i[0])
        print(new_df.iloc[i[0]].title)
    return movies_list
```

```python
recommend('Avatar')
```
